```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = (20, 6)
plt.rcParams['font.size'] = 14
```

```python
# plotting sin(x)
x = np.linspace(0, 2* np.pi, 100)
y = np.sin(x)
```

```python
plt.scatter(x, y)
```

```python
from sklearn import tree
```

```python
# decision tree giving avg for [0:pi] and [pi:2pi]
regression = tree.DecisionTreeRegressor(max_depth=1)
regression.fit(x.reshape(-1, 1), y)

yp = regression.predict(x.reshape(-1,1))

plt.scatter(x, yp)
```

```python
# testing model for f(2)
regression.predict([[2]])
```

```python
path = regression.decision_path(x.reshape(-1, 1))
```

```python
path.todense()
```

## Decision tree
- Creates decision tree with only one node (see above where max_depth = 1).  Asks if value is <= to 3.142.  If yes, value = 0.63.  If no, value = -0.63

```python
tree.plot_tree(regression)
```

Now lets try with more nodes.

```python
regression = tree.DecisionTreeRegressor(max_depth=2)
regression.fit(x.reshape(-1, 1), y)

yp = regression.predict(x.reshape(-1,1))

plt.scatter(x, yp)
plt.plot(x,y)
```

Two nodes (max_depth=2) now gives four values. This gets closer to the sin(x) shape

```python
tree.plot_tree(regression)
```

```python
regression = tree.DecisionTreeRegressor(max_depth=10)
regression.fit(x.reshape(-1, 1), y)

yp = regression.predict(x.reshape(-1,1))

plt.scatter(x, yp)
plt.plot(x,y)
```

Now add some random noise

```python
x = np.linspace(0, 2* np.pi, 100)
y = np.sin(x) + .5*np.random.random(100)
```

```python
plt.scatter(x, y)
```

Remember to not overfit.  Model should reflect general shape and not account for noise.

```python
regression = tree.DecisionTreeRegressor(max_depth=8)
regression.fit(x.reshape(-1, 1), y)

yp = regression.predict(x.reshape(-1,1))

plt.scatter(x, yp)
plt.plot(x,y)
```

```python
regression = tree.DecisionTreeRegressor(max_depth=8, min_samples_split=8)
regression.fit(x.reshape(-1, 1), y)

yp = regression.predict(x.reshape(-1,1))

plt.scatter(x, yp)
plt.plot(x, y)
```

```python
tree.plot_tree(regression)
```
